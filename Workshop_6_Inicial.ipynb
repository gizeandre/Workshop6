```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patheffects as path_effects
import contextily as cx
from pathlib import Path
import plotly.express as px
```

```python
Ruta_Carpeta=r"/notebooks/Datos"

# Municipios Colombia
shp_path=Ruta_Carpeta+r"/municipios_colombia.shp"
muni = gpd.read_file(shp_path)
muni_3857 = muni.to_crs(3857)
muni_9377= muni.to_crs(epsg=9377)
print(muni.head())

#Evaluaciones Agropecuarias Municipales EVA
EVA_csv_path  = Ruta_Carpeta+"/Evaluaciones_Agropecuarias_Municipales_EVA.csv"
eva = pd.read_csv(EVA_csv_path, encoding='utf-8')
print(eva.head())

#Estaciones IDEAM
Estaciones_csv_path=Ruta_Carpeta+"/Catalogo_Estaciones_IDEAM.csv"
estaciones = pd.read_csv(Estaciones_csv_path, encoding='utf-8')

#Extraer coordenadas de ubicación
coordenadas = estaciones['Ubicación'].str.strip('()').str.split(',', expand=True)
coordenadas['y'] = coordenadas[0].astype(float)  #se extraen por separado para eliminar espacios
coordenadas['x'] = coordenadas[1].astype(float)

# Crear geometría
coordenadas_geom=gpd.points_from_xy(coordenadas['x'], coordenadas['y'])
estaciones_gdf = gpd.GeoDataFrame(estaciones, geometry=coordenadas_geom)
estaciones_gdf.set_crs(epsg=4326, inplace=True)
print("CRS Inicial Estaciones:", estaciones_gdf.crs)  

#Proyección coordenadas
estaciones_gdf_9377= estaciones_gdf.to_crs(epsg=9377)
estaciones_gdf_3857= estaciones_gdf.to_crs(epsg=3857)

#Capa Precipitación
root_folder=r"/notebooks/Datos/"
DATA = Path(root_folder+'precipitacion.csv')
print (DATA)

#Verificación espacial mediante visualización
fig, ax = plt.subplots(figsize=(10, 10))
estaciones_gdf_3857.plot(ax=ax, color='red', markersize=20, alpha=0.7, label='Estaciones')
muni_3857.boundary.plot(ax=ax, edgecolor='black', linewidth=0.2)

cx.add_basemap(ax, source=cx.providers.CartoDB.Voyager)
ax.set_title("Ubicación de Estaciones Meteorológicas", fontsize=14)
ax.set_axis_off()
ax.legend()
plt.tight_layout()
plt.show()
```
